```python
#Matrix operations
```

```python
A=np.arange(1,5).reshape(2,2)
B=np.arange(5,9).reshape(2,2)
```

```python
A@B  # @ for matrix multiplication
```

```python
np.dot(A,B)    # Function for array multiplication
```

```python
A.T         # .T for transposing an array
```

```python
#Stacking Array
```

```python
l1=np.arange(1,6)
l2=np.arange(2,7)
l3=np.arange(3,8)
np.vstack((l1,l2,l3))
```

```python
np.hstack((l1,l2,l3))
```

```python
np.column_stack((l1,l2))
```
