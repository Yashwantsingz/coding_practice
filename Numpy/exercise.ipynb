```python
import numpy as np
```

```python
s=np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
s
```

```python
sum_row=np.sum(s,axis=1)
sum_column=np.sum(s,axis=0)
for i in range(len(sum_row)):
    if sum_row[i]!=45 or sum_column[i]!=45:
        print("Sudoku is not valid")
        break
else:
    print("Sudoku is valid")
```

```python
for i in range(0,9,3):
    for j in range(0,7,3):
        if s[i:i+3,j:j+3].sum() !=45:
            print("Sudoku is not valid")
            break
else:
    print("Sudoku is valid")
```
