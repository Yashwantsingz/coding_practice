```python
import numpy as np
import pandas as pd
```

```python
data={
    "A":[1,2,np.nan,4,5],
    "B":[1,2,3,4,5],
    "C":[1,2,3,np.nan,np.nan],
    "D":[1,np.nan,np.nan,np.nan,5]
}
df=pd.DataFrame(data)
df
```

```python
df.isna() #tell in df where the NaN(null value) exists
```

```python
df.isna().sum() #tells total no. of null values in each column
```

```python
#Removing the missing data
```

```python
df.dropna() #removes all the rows having NaN
```

```python
df.dropna(thresh=3) #checks if atleast there are non NaN values equal to thresh
```

```python
#Filling the missing data
```

```python
df.fillna(34)
```

```python
values={'A':10,'B':20,'C':30,'D':40}
df.fillna(value=values)
```

```python
df.fillna(df.mean(),inplace=True)
```
