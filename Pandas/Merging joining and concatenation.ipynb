```python
import numpy as np
import pandas as pd
```

```python
#Merging 2 dataframes
```

```python
employees=pd.DataFrame ({
'employee_id': [1,2,3,4,5],
"name": ['John','Anna',"Peter", "Linda","Bob"],
'department': ['HR', 'IT', 'Finance', 'IT',"HR"]
})
# DataFrame 2: Salary information I
salaries=pd.DataFrame({
'employee_id': [1, 2, 3, 6, 7],
'salary': [60000, 80000, 65000,70000,90000],
'bonus':[5000,10000,7000,8000,12000]})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner') #takes only common employee_id from both dataframes
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer') #print details of every emp_id if any missing takes null there
```

```python
pd.merge(employees,salaries,on="employee_id",how='left') #prints all the details of all emp_id of left DF
```

```python
pd.merge(employees,salaries,on="employee_id",how='right') #prints all the details of all emp_id of right DF
```

```python
#Concatenation of 2 dataframes
```

```python
df1=pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})
df2=pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
pd.concat([df1,df2]) #concatenates on the bases of columns
```

```python
pd.concat([df1,df2],axis=1) #concatenates on the bases of rows
```

```python
# Joining Databases
```

```python
df1=pd. DataFrame({
    'name':['Alice','Bob','Charlie']}
    ,index=[1, 2, 3])
# Second DataFrame
df2=pd. DataFrame ({
    'score': [85, 90, 75]
    },index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
